```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web 
import matplotlib.pyplot as plt
```

```python
tickers = ['^OMX', '^IXIC', '^GSPC', '^BSESN']
```

I have chosen Stockexchanges from 
Sweden: Nasdaq OMX 30
USA: Nasdaq composite 
USA: S&P 500
India: BSE
it must be noted that the ticker symbol starts with the hat sign '^' and is given with the name of index when you search for respective index in Yahoo finance. Additionally, I was unable to get historical data from FTSE 100 from UK - it gave error. 

```python
data = pd.DataFrame()
for t in tickers:
    data[t] = web.DataReader(t, data_source = 'yahoo', start = '2000-1-1')['Adj Close']
```

```python
data.head()
```

```python
data.tail()
```

```python
(data / data.iloc[0] * 100).plot(figsize=(15,6))
plt.show
```

the above graph shows how the NASDAQ composite index is the clear winner. The second rise in the stock index has been for the Indian - Bombay stock exchange as it may be attributed to the emerging economy. The worst performer had been the Swedish index NASDAQ OMX. 

```python
annual_returns = ((data / data.shift(1)) - 1).mean() * 250 
annual_returns
```
