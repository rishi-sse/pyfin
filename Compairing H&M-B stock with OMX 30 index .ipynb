```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web 
import matplotlib.pyplot as plt
```

```python
tickers = ['^OMX', 'HM-B.ST']
```

```python
data = pd.DataFrame()
for t in tickers:
    data[t] = web.DataReader(t, data_source = 'yahoo', start = '2000-1-1')['Adj Close']
```

```python
data.head()
```

# Without Normalization 

```python
data.plot(figsize = (15,6))
plt.show()
```

# After Normalization 

```python
(data / data.iloc[0] * 100).plot(figsize = (15,6))
plt.show()
```

# Returns

```python
annual_returns = (data / data.shift(1) - 1).mean() * 250
annual_returns
```

```python
log_returns = np.log(data / data.shift(1)).mean() * 250
log_returns
```
